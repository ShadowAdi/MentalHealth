```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier,ExtraTreesClassifier,HistGradientBoostingClassifier,GradientBoostingClassifier
from sklearn.tree  import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,accuracy_score,precision_score,f1_score,recall_score
```

```python
def modelTrainer(model):
    np.random.seed(32)
    modelName=model()
    modelName.fit(X_train,y_train)
    modelPreds=modelName.predict(X_test)
    return {
        "ModelName":model,
        "Precision Score":precision_score(y_test,modelPreds),
        "Accuracy Score":accuracy_score(y_test,modelPreds),
        "F1 Score":f1_score(y_test,modelPreds),
        "Recall Score":recall_score(y_test,modelPreds),
    }
```

# **A/C To Accuracy Score-Ada Boost Is The Best Model**

### We Will Do GridSearchCV on at least 3 Models

```python
from sklearn.model_selection import GridSearchCV
```

```python
ada_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5]
}
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
ada_grid = GridSearchCV(AdaBoostClassifier(), param_grid=ada_params, cv=5,scoring="accuracy")
ada_grid.fit(X_train, y_train)
print("Best parameters for AdaBoost:", ada_grid.best_params_)
```

```python
print("Best Estimator for AdaBoost:", ada_grid.best_estimator_)
```

```python
print("Best Score for AdaBoost:", ada_grid.best_score_)
```

```python
hist_params = {
    'max_iter': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5]
}
hist_grid = GridSearchCV(HistGradientBoostingClassifier(), param_grid=hist_params, cv=3)
hist_grid.fit(X_train, y_train)
print("Best parameters for HistGradientBoosting:", hist_grid.best_params_)
print("Best Score for HistGradientBoosting:", hist_grid.best_score_)
```

```python
grad_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5],
    'max_depth': [3, 5, 7]
}
grad_grid = GridSearchCV(GradientBoostingClassifier(), param_grid=grad_params, cv=3)
grad_grid.fit(X_train, y_train)
print("Best parameters for GradientBoosting:", grad_grid.best_params_)
print("Best Estimator for GradientBoosting:", grad_grid.best_estimator_)
print("Best Score for GradientBoosting:", grad_grid.best_score_)

```

```python
xgb_params = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5],
    'max_depth': [3, 5, 7]
}
xgb_grid = GridSearchCV(XGBClassifier(), param_grid=xgb_params, cv=3)
xgb_grid.fit(X_train, y_train)
print("Best parameters for XGB:", xgb_grid.best_params_)
print("Best Estimator for XGB:", xgb_grid.best_estimator_)
print("Best Score for XGB:", xgb_grid.best_score_)
```
